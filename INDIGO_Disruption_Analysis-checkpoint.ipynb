```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
np.random.seed(42)

dates = pd.date_range("2025-06-01", "2025-06-10")
airports = ["DEL", "BOM", "BLR", "HYD", "CCU"]
```

```python
data = []

for date in dates:
    for i in range(50):  # 50 - 60 flights per day
        origin = np.random.choice(airports)
        destination = np.random.choice([a for a in airports if a != origin])
        
        if date <= pd.Timestamp("2025-06-03"):
            phase = "Pre"
            delay = np.random.normal(5, 5)
            status = "On-Time" if delay < 15 else "Delayed"
        elif date <= pd.Timestamp("2025-06-06"):
            phase = "Disruption"
            delay = np.random.normal(70, 40)
            status = "Cancelled" if np.random.rand() < 0.25 else "Delayed"
        else:
            phase = "Recovery"
            delay = np.random.normal(30, 15)
            status = "Delayed" if delay > 15 else "On-Time"

        delay = max(0, int(delay))

        data.append([
            date,
            "IndiGo",
            origin,
            destination,
            delay,
            status,
            phase,
            180
        ])
```

```python
df = pd.DataFrame(
    data,
    columns=[
        "flight_date",
        "airline",
        "origin_airport",
        "destination_airport",
        "departure_delay_min",
        "flight_status",
        "disruption_phase",
        "estimated_passengers"
    ]
)

df.head()
```

```python
df = pd.read_csv("../data/flight_operations_data.csv")
df.head()
```

```python
df["flight_date"] = pd.to_datetime(df["flight_date"])
```

```python
df.dtypes
```

```python
df.loc[df["flight_status"] == "Cancelled", "departure_delay_min"] = 0
```

```python
def delay_category(delay):
    if delay == 0:
        return "On-Time"
    elif delay <= 30:
        return "Low Delay"
    elif delay <= 60:
        return "Medium Delay"
    else:
        return "High Delay"

df["delay_category"] = df["departure_delay_min"].apply(delay_category)
```

```python
df["disruption_flag"] = df["disruption_phase"].apply(
    lambda x: 1 if x == "Disruption" else 0
)
```

```python
df["is_cancelled"] = np.where(df["flight_status"] == "Cancelled", 1, 0)
```

```python
def passenger_impact(row):
    if row["flight_status"] == "Cancelled":
        return row["estimated_passengers"]
    elif row["departure_delay_min"] > 60:
        return row["estimated_passengers"] * 0.7
    elif row["departure_delay_min"] > 30:
        return row["estimated_passengers"] * 0.4
    elif row["departure_delay_min"] > 0:
        return row["estimated_passengers"] * 0.1
    else:
        return 0

df["passenger_impact"] = df.apply(passenger_impact, axis=1)
```

```python
df["disruption_score"] = (
    df["departure_delay_min"] +
    (df["is_cancelled"] * 180)
)
```

```python
total_flights = len(df)

delay_rate = (df["flight_status"] == "Delayed").mean() * 100
cancel_rate = (df["flight_status"] == "Cancelled").mean() * 100

mean_delay = df[df["departure_delay_min"] > 0]["departure_delay_min"].mean()

total_passenger_impact = df["passenger_impact"].sum()

total_flights, delay_rate, cancel_rate, mean_delay, total_passenger_impact
```

```python
phase_kpis = df.groupby("disruption_phase").agg(
    flights=("flight_status", "count"),
    avg_delay=("departure_delay_min", "mean"),
    cancel_rate=("is_cancelled", "mean"),
    total_passenger_impact=("passenger_impact", "sum"),
    avg_disruption_score=("disruption_score", "mean")
)

phase_kpis
```

```python
daily_dsi = df.groupby("flight_date")["disruption_score"].sum()
daily_dsi
```

```python
airport_delay = (
    df.groupby("origin_airport")["departure_delay_min"]
    .mean()
    .sort_values(ascending=False)
)
airport_delay
```

```python
airport_dsi = (
    df.groupby("origin_airport")["disruption_score"]
    .sum()
    .sort_values(ascending=False)
)
airport_dsi
```

```python
baseline_dsi = df[df["disruption_phase"] == "Pre"].groupby("flight_date")["disruption_score"].sum().mean()
baseline_dsi
```

```python
recovery_date = daily_dsi[daily_dsi <= baseline_dsi * 1.2].index.min()
recovery_date
```

```python
incident_start = df[df["disruption_phase"] == "Disruption"]["flight_date"].min()
mttr_days = (recovery_date - incident_start).days
mttr_days
```
